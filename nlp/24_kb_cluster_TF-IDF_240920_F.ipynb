```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from konlpy.tag import Okt
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np

# 텍스트 파일 경로 설정
text_file_path = 'data/Q1_2.txt'

# 한글 폰트 경로 설정
import matplotlib.font_manager as fm
font_path = 'C:/Windows/Fonts/malgun.ttf'  # 시스템에 맞게 변경
fontprop = fm.FontProperties(fname=font_path)
plt.rc('font', family='Malgun Gothic')

# 1. 데이터 로드
with open(text_file_path, 'r', encoding='utf-8') as file:
    sentences = file.readlines()

# 2. 전처리
sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

# 불필요한 단어들 제거
stopwords = ['없음', '모름', '없다', '기억안남', '모르겠다', '없어요', '생각안남', '없습니다',
             '특별히', '이름이', '감사합니다']
filtered_sentences = [sentence for sentence in sentences if sentence.lower() not in stopwords]

# 3. 명사 추출을 위한 Okt 객체 생성
okt = Okt()

def extract_nouns(text):
    nouns = okt.nouns(text)
    return ' '.join(nouns)

# 문장마다 명사만 추출
nouns_sentences = []
for sentence in filtered_sentences:
    nouns = extract_nouns(sentence)
    if nouns.strip():  # 명사 추출 결과가 빈 문자열이 아닌 경우에만 추가
        nouns_sentences.append(nouns)
    else:
        nouns_sentences.append('')  # 빈 문자열도 유지하여 인덱스 일치

# 빈 문자열만으로 구성된 문장 제거
filtered_data = [(nouns, sentence) for nouns, sentence in zip(nouns_sentences, filtered_sentences) if nouns.strip()]
if filtered_data:
    nouns_sentences, filtered_sentences = zip(*filtered_data)
    nouns_sentences = list(nouns_sentences)
    filtered_sentences = list(filtered_sentences)
else:
    print("유효한 데이터가 없습니다.")
    exit()

# 4. TF-IDF 벡터화 (명사에 대해서만)
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(nouns_sentences)

# 5. 초기 클러스터 수 추천 (실루엣 점수 기반)
print("\n초기 클러스터 수를 추천하기 위해 실루엣 점수를 계산합니다...")
possible_cluster_numbers = range(2, min(21, len(filtered_sentences)))
silhouette_scores = []

for n_clusters in possible_cluster_numbers:
    if n_clusters >= X.shape[0]:
        break
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(X)
    if len(set(labels)) > 1:
        score = silhouette_score(X, labels)
        silhouette_scores.append((n_clusters, score))
    else:
        silhouette_scores.append((n_clusters, -1))

# 가장 높은 실루엣 점수를 가지는 클러스터 수 찾기
best_n_clusters = max(silhouette_scores, key=lambda x: x[1])[0]
print("\n실루엣 점수 기반 추천 초기 클러스터 수:")
for n_clusters, score in silhouette_scores:
    print(f" - 클러스터 수: {n_clusters}, 실루엣 점수: {score:.4f}")
print(f">>> 추천 클러스터 수: {best_n_clusters}")

# 6. 초기 클러스터 수 입력
while True:
    try:
        num_clusters = int(input(f"\n초기 클러스터 수를 입력하세요 (2~{len(filtered_sentences)}, 추천: {best_n_clusters}): "))
        if num_clusters < 2 or num_clusters > len(filtered_sentences):
            print(f"클러스터 수는 2에서 {len(filtered_sentences)} 사이의 정수여야 합니다.")
            continue
        break
    except ValueError:
        print("유효한 숫자를 입력해주세요.")

# 7. 클러스터링
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# 8. 각 클러스터의 대표 문장 및 핵심 키워드 추출
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()

clustered_sentences = [[] for _ in range(num_clusters)]
clustered_nouns = [[] for _ in range(num_clusters)]  # 각 클러스터의 명사 문장
for i, label in enumerate(kmeans.labels_):
    clustered_sentences[label].append(filtered_sentences[i])
    clustered_nouns[label].append(nouns_sentences[i])

keywords = []
for i in range(num_clusters):
    # 각 클러스터에서 TF-IDF 가중치가 가장 높은 단어(명사)를 추출
    key_term = terms[order_centroids[i, 0]]
    keywords.append(key_term)

# 클러스터를 딕셔너리 형태로 저장
clusters = []
for i in range(num_clusters):
    cluster = {
        'cluster_id': str(i),
        'sentences': clustered_sentences[i],
        'nouns_sentences': clustered_nouns[i],  # 추가된 부분
        'keyword': keywords[i],
        'size': len(clustered_sentences[i]),
    }
    clusters.append(cluster)

# 클러스터 정보 출력 및 데이터 수집 함수
def print_and_collect_cluster_info(clusters):
    sorted_clusters = sorted(clusters, key=lambda x: x['size'], reverse=True)
    excel_data = []
    summary_data = []
    keyword_data = []
    for cluster in sorted_clusters:
        cluster_id = cluster['cluster_id']
        freq = cluster['size']
        key = cluster['keyword']
        sentences = cluster['sentences']
        nouns_sentences_cluster = cluster['nouns_sentences']  # 수정된 부분

        # 요약 문장 생성
        if len(sentences) > 0:
            # 빈 문자열 제거
            filtered_cluster_data = [(nouns, sentence) for nouns, sentence in zip(nouns_sentences_cluster, sentences) if nouns.strip()]
            if filtered_cluster_data:
                nouns_sentences_cluster, sentences = zip(*filtered_cluster_data)
                nouns_sentences_cluster = list(nouns_sentences_cluster)
                sentences = list(sentences)
                # TF-IDF 벡터화
                vectorizer_cluster = TfidfVectorizer()
                X_cluster = vectorizer_cluster.fit_transform(nouns_sentences_cluster)
                # 클러스터 중심 계산
                centroid = X_cluster.mean(axis=0)
                # 각 문장과 중심 간의 거리 계산
                distances = np.linalg.norm(X_cluster - centroid, axis=1)
                # 가장 가까운 문장 선택
                closest_index = distances.argmin()
                summary_sentence = sentences[closest_index]
            else:
                summary_sentence = "No summary available"
        else:
            summary_sentence = "No summary available"

        print(f"\nCluster {cluster_id} (빈도수: {freq}):")
        print(f"Keyword: {key}")
        print(f"Summary: {summary_sentence}")
        print("Sentences:")
        excel_data.append([f"Cluster {cluster_id} (빈도수: {freq})", f"Keyword: {key}",
                           f"Summary: {summary_sentence}"])
        summary_data.append((summary_sentence, freq))
        keyword_data.append((key, freq))
        for sentence in sentences:
            print(f"- {sentence}")
            excel_data.append(["", "", f"- {sentence}"])
        excel_data.append(["", "", ""])
    return excel_data, summary_data, keyword_data

# 초기 클러스터 정보 출력 및 데이터 수집
excel_data, summary_data, keyword_data = print_and_collect_cluster_info(clusters)

# 9. 결과 저장

# 저장 폴더 경로 설정
save_folder = os.path.join('data', os.path.splitext(os.path.basename(text_file_path))[0])

# 폴더가 없는 경우 생성
os.makedirs(save_folder, exist_ok=True)

# 엑셀 파일 경로 설정
excel_file_path = os.path.join(save_folder, 'clustering_results.xlsx')

# 엑셀 파일로 저장
df = pd.DataFrame(excel_data, columns=["", "", ""])
df.to_excel(excel_file_path, index=False, header=False)

print(f"\nClustering results saved to {excel_file_path}")

# 10. Keyword를 변수로 하여 가로 막대 그래프 생성 (빈도수가 높은 순으로 정렬)
def draw_keyword_frequency_graph(keyword_data, graph_file_path):
    if keyword_data:
        try:
            keywords_list, frequencies = zip(*sorted(keyword_data, key=lambda x: x[1], reverse=True))
        except ValueError:
            # keyword_data가 비어있을 경우 처리
            keywords_list, frequencies = [], []

        if keywords_list and frequencies:
            # 그래프 크기 동적 조절 (클러스터 수에 따라)
            num_keywords = len(keywords_list)
            height_per_bar = 0.5  # 각 바의 높이
            margin = 2  # 상하 여백
            figure_height = min(max(6, num_keywords * height_per_bar + margin), 20)  # 최소 6인치, 최대 20인치

            plt.figure(figsize=(12, figure_height))  # 가로는 고정, 세로는 동적으로 조절
            bars = plt.barh(keywords_list, frequencies, color='skyblue')
            plt.gca().invert_yaxis()  # 빈도수가 높은 것이 위로 오도록 설정

            plt.xlabel('Frequency')
            plt.title('Cluster Keyword Frequency')

            # **여기에서 빈도수 숫자를 막대 그래프에 표시합니다.**
            for bar, freq in zip(bars, frequencies):
                plt.text(freq + 0.5, bar.get_y() + bar.get_height()/2, str(freq), va='center')

            plt.tight_layout()

            # 그래프 저장
            plt.savefig(graph_file_path)
            plt.show()

            print(f"Cluster keyword frequency graph saved to {graph_file_path}")
        else:
            print("클러스터 데이터가 없어 그래프를 생성할 수 없습니다.")
    else:
        print("클러스터 데이터가 없어 그래프를 생성할 수 없습니다.")

# 그래프 파일 경로 설정
graph_file_path = os.path.join(save_folder, 'cluster_keyword_frequency.png')

# 그래프 그리기
draw_keyword_frequency_graph(keyword_data, graph_file_path)

# 11. 재클러스터링 루프
while True:
    user_input = input("\n재클러스터링할 클러스터의 Keyword를 입력하세요 (종료하려면 '끝' 입력): ")
    if user_input.strip() == '끝':
        break
    else:
        # 입력된 키워드에 해당하는 클러스터 찾기
        matching_clusters = [cluster for cluster in clusters if user_input == cluster['keyword']]
        if not matching_clusters:
            print(f"Keyword '{user_input}'에 해당하는 클러스터가 없습니다. 다시 입력해주세요.")
        else:
            # 해당 클러스터들에 대해 재클러스터링 수행
            for cluster in matching_clusters:
                sentences_to_recluster = cluster['sentences']
                nouns_sentences_to_recluster = cluster['nouns_sentences']
                cluster_size = cluster['size']
                if cluster_size <= 1:
                    print(f"클러스터 '{cluster['keyword']}'에 문장이 하나이므로 재클러스터링을 수행할 수 없습니다.")
                    continue
                else:
                    # 가능한 클러스터 수 범위 설정 (2부터 cluster_size까지)
                    possible_cluster_numbers = range(2, cluster_size+1)
                    silhouette_scores = []

                    # 실루엣 점수 계산
                    # 빈 문자열 제거
                    filtered_recluster_data = [(nouns, sentence) for nouns, sentence in zip(nouns_sentences_to_recluster, sentences_to_recluster) if nouns.strip()]
                    if not filtered_recluster_data:
                        print(f"클러스터 '{cluster['keyword']}'에서 유효한 명사를 추출할 수 없습니다.")
                        continue
                    nouns_sentences_to_recluster, sentences_to_recluster = zip(*filtered_recluster_data)
                    nouns_sentences_to_recluster = list(nouns_sentences_to_recluster)
                    sentences_to_recluster = list(sentences_to_recluster)
                    vectorizer_recluster = TfidfVectorizer()
                    X_recluster = vectorizer_recluster.fit_transform(nouns_sentences_to_recluster)

                    for n_clusters in possible_cluster_numbers:
                        if n_clusters >= X_recluster.shape[0]:
                            break
                        kmeans_recluster = KMeans(n_clusters=n_clusters, random_state=42)
                        labels = kmeans_recluster.fit_predict(X_recluster)
                        if len(set(labels)) > 1:
                            score = silhouette_score(X_recluster, labels)
                            silhouette_scores.append((n_clusters, score))
                        else:
                            silhouette_scores.append((n_clusters, -1))

                    if silhouette_scores:
                        # 가장 높은 실루엣 점수를 가지는 클러스터 수 찾기
                        best_n_clusters = max(silhouette_scores, key=lambda x: x[1])[0]
                        print("실루엣 점수 기반 추천 재클러스터링 수:")
                        for n_clusters, score in silhouette_scores:
                            print(f" - 클러스터 수: {n_clusters}, 실루엣 점수: {score:.4f}")
                        print(f">>> 추천 클러스터 수: {best_n_clusters}")
                    else:
                        best_n_clusters = 2
                        print("실루엣 점수를 계산할 수 없어 기본값으로 클러스터 수 2를 추천합니다.")

                    # 새로운 클러스터 수 입력
                    while True:
                        try:
                            recluster_num = int(input(
                                f"클러스터 '{cluster['keyword']}'에 대한 재클러스터링 수를 입력하세요 "
                                f"(2~{cluster_size}, 추천: {best_n_clusters}): "))
                            if recluster_num < 2 or recluster_num > cluster_size:
                                print(f"유효하지 않은 클러스터 수입니다. 2에서 {cluster_size} 사이의 값을 입력해주세요.")
                                continue
                            break
                        except ValueError:
                            print("유효한 숫자를 입력해주세요.")

                    # 재클러스터링 수행
                    kmeans_recluster = KMeans(n_clusters=recluster_num, random_state=42)
                    kmeans_recluster.fit(X_recluster)
                    labels = kmeans_recluster.labels_
                    order_centroids_recluster = kmeans_recluster.cluster_centers_.argsort()[:, ::-1]
                    terms_recluster = vectorizer_recluster.get_feature_names_out()
                    new_clustered_sentences = [[] for _ in range(recluster_num)]
                    new_clustered_nouns = [[] for _ in range(recluster_num)]
                    for i, label in enumerate(labels):
                        new_clustered_sentences[label].append(sentences_to_recluster[i])
                        new_clustered_nouns[label].append(nouns_sentences_to_recluster[i])
                    new_keywords = []
                    for i in range(recluster_num):
                        key_term = terms_recluster[order_centroids_recluster[i, 0]]
                        new_keywords.append(key_term)
                    # 원래 클러스터 제거
                    clusters.remove(cluster)
                    # 새로운 클러스터 추가
                    existing_ids = [c['cluster_id'] for c in clusters]
                    if existing_ids:
                        numeric_ids = []
                        for cid in existing_ids:
                            try:
                                numeric_part = int(cid.split('-')[0])
                                numeric_ids.append(numeric_part)
                            except ValueError:
                                pass
                        if numeric_ids:
                            next_cluster_id = max(numeric_ids) + 1
                        else:
                            next_cluster_id = 0
                    else:
                        next_cluster_id = 0
                    for i in range(recluster_num):
                        new_cluster_id = f"{next_cluster_id}-{i}"
                        new_cluster = {
                            'cluster_id': new_cluster_id,
                            'sentences': new_clustered_sentences[i],
                            'nouns_sentences': new_clustered_nouns[i],
                            'keyword': new_keywords[i],
                            'size': len(new_clustered_sentences[i]),
                        }
                        clusters.append(new_cluster)
            # 재클러스터링 후 클러스터 정보 출력 및 데이터 수집
            excel_data, summary_data, keyword_data = print_and_collect_cluster_info(clusters)

            # 결과 저장 및 그래프 업데이트
            # 엑셀 파일로 저장
            df = pd.DataFrame(excel_data, columns=["", "", ""])
            df.to_excel(excel_file_path, index=False, header=False)

            # 그래프 업데이트
            draw_keyword_frequency_graph(keyword_data, graph_file_path)
```
