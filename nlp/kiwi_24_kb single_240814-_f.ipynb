```python
# 라이브러리 불러오기
import os
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd  # 추가: pandas 라이브러리 불러오기

# 한글 폰트 설정
font_path = 'C:/Windows/Fonts/malgun.ttf'  # 맑은 고딕 폰트 경로
fontprop = fm.FontProperties(fname=font_path)
plt.rc('font', family=fontprop.get_name())

# 텍스트 파일 불러오기(파일명만 변경)
text_file_path = 'data/q25c.txt'
with open(text_file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# 새로운 폴더 생성
file_name = os.path.basename(text_file_path).split('.')[0]
output_dir = f'data/{file_name}'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 파일에서 읽어온 텍스트를 줄 단위로 분리하여 명사 리스트로 처리
nouns = text.splitlines()

# 명사와 빈도수 계산
nouns_counter = Counter(nouns)

# 빈도수가 2개 이상인 명사만 추출하여 딕셔너리 생성
nouns_dict = {noun: count for noun, count in nouns_counter.items() if count >= 2}

# 워드클라우드 시각화 1
wordcloud = WordCloud(
    font_path=font_path,
    background_color='white',
    width=800,
    height=800,
    margin=20
).generate_from_frequencies(nouns_dict)

plt.figure(figsize=(3.15, 3.15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.savefig(
    f'{output_dir}/wordcloud(800X800).png',
    dpi=254,
    bbox_inches='tight',
    pad_inches=0
)
plt.show()

# 워드클라우드 시각화 2
wordcloud = WordCloud(
    font_path=font_path,
    background_color='white',
    width=1200,
    height=800,
    margin=20
).generate_from_frequencies(nouns_dict)

plt.figure(figsize=(4.72, 3.15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.savefig(
    f'{output_dir}/wordcloud(1200X800).png',
    dpi=254,
    bbox_inches='tight',
    pad_inches=0
)
plt.show()

# 빈도수 데이터 엑셀로 저장
# nouns_dict를 데이터프레임으로 변환하고 빈도수 기준으로 내림차순 정렬
df = pd.DataFrame(list(nouns_dict.items()), columns=['Word', 'Frequency'])
df = df.sort_values(by='Frequency', ascending=False)  # 빈도수 내림차순 정렬

# 엑셀 파일로 저장
excel_path = f'{output_dir}/wordcloud_data.xlsx'
df.to_excel(excel_path, index=False)
```

# 워드클라우드 글자 키우기

```python
# 라이브러리 불러오기
import os
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
font_path = 'C:/Windows/Fonts/malgun.ttf'  # 맑은 고딕 폰트 경로
fontprop = fm.FontProperties(fname=font_path)
plt.rc('font', family=fontprop.get_name())

# 텍스트 파일 불러오기(파일명만 변경)
text_file_path = 'data/Q25C.txt'
with open(text_file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# 새로운 폴더 생성
file_name = os.path.basename(text_file_path).split('.')[0]
output_dir = f'data/{file_name}'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 파일에서 읽어온 텍스트를 줄 단위로 분리하여 명사 리스트로 처리
nouns = text.splitlines()

# 명사와 빈도수 계산
nouns_counter = Counter(nouns)

# 상위 100개의 명사 추출 및 빈도수 바탕 딕셔너리 생성 (추출 단어 수 증가)
most_common_nouns = nouns_counter.most_common(100)
nouns_dict = dict(most_common_nouns)

# 워드클라우드 시각화
wordcloud = WordCloud(
    font_path=font_path,
    background_color='white',
    width=800,
    height=800,
    margin=10,
    max_font_size=250,  # 최대 폰트 크기를 키움
    relative_scaling=0.5,  # 폰트 크기 간 상대적 차이를 줄임
    scale=1.5  # 전체적으로 크기를 키움
).generate_from_frequencies(nouns_dict)

plt.figure(figsize=(8, 8))  # 전체 크기를 키움
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.savefig(
    f'{output_dir}/wordcloud(800X800)_scaled.png',
    dpi=254,
    bbox_inches='tight',
    pad_inches=0
)
plt.show()



# 워드클라우드 시각화 2
wordcloud = WordCloud(
    font_path=font_path,
    background_color='white',
    width=1200,
    height=800,
    margin=10,
    max_font_size=250,  # 최대 폰트 크기를 키움
    relative_scaling=0.5,  # 폰트 크기 간 상대적 차이를 줄임
    scale=1.5  # 전체적으로 크기를 키움
).generate_from_frequencies(nouns_dict)

plt.figure(figsize=(12, 8))  # 전체 크기를 키움
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.savefig(
    f'{output_dir}/wordcloud(1200X800)_scaled.png',
    dpi=254,
    bbox_inches='tight',
    pad_inches=0
)
plt.show()
```
